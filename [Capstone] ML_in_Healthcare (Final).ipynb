### **ECG5000**  
> The original data set for *ECG5000* is a 20-hour long ECG downloaded from Physionet_ATM [1]. The database is BIDMC Congestive Heart Failure Database(chfdb) and the record is chf07. It was originally published by Goldberger et al. [2]. The data were pre-processed in two steps, first extracting each heartbeat and then making each heartbeat equal length using interpolation. The class values were obtained by automated annotation.

## TODO 2 Preprocessing (4 points)

1. Normalize each sample (features) following the time-dimention, using the *sklearn.preprocessing.StandardScaler* function. Please remember to install and import the *sklearn* library.

2. Note, as the training and validation sets are already splitted into two different files, you need to do the preprocessing for them independently (seperately).

```python
# SNIPPET 1: Preprocessing training and validation data
# Separate labels and features for training data
X_train = np.array([sample[1:] for sample in ECGdataset_train])  # Features (all except first column)
y_train = np.array([sample[0] for sample in ECGdataset_train])   # Labels (first column)

# Separate labels and features for validation data
X_val = np.array([sample[1:] for sample in ECGdataset_validation])  # Features
y_val = np.array([sample[0] for sample in ECGdataset_validation])   # Labels

print(f"Training feature shape: {X_train.shape}, Training label shape: {y_train.shape}")
print(f"Validation feature shape: {X_val.shape}, Validation label shape: {y_val.shape}")

# Normalize features using StandardScaler
scaler = StandardScaler()
X_train_normalized = scaler.fit_transform(X_train)
# Use the same scaler for validation data (only transform, not fit_transform)
X_val_normalized = scaler.transform(X_val)

# Visualize the first sample's feature after normalization
plt.figure(figsize=(10, 4))
plt.plot(X_train_normalized[0])
plt.title('First Sample ECG Signal (Normalized)')
plt.xlabel('Time')
plt.ylabel('Normalized Amplitude')
plt.show()
```

```python
# Assuming ECGdataset_train is already defined and has the same structure
# Separate features
ecg_feature = np.array([sample[1:] for sample in ECGdataset_train])  # Features

# Normalize features using StandardScaler
scaler = StandardScaler()
ecg_data_normalized = scaler.fit_transform(ecg_feature)

# Visualize the first sample's feature after normalization
plt.figure(figsize=(10, 4))
plt.plot(ecg_data_normalized[0])
plt.title('First Sample ECG Signal (Normalized)')
plt.xlabel('Time')
plt.ylabel('Normalized Amplitude')
plt.show()
```

## TODO 4 Train a KNN classifier (4 points)

1. Use Sklearn library, train a KNN classifier (set K equals to *5*), feed the *Training_set* to the classifier.

```python
# Initialize KNN classifier with K=5
knn_classifier = KNeighborsClassifier(n_neighbors=5)

# Train the classifier on the normalized training data
knn_classifier.fit(X_train_normalized, y_train)
```

## TODO 5 Validate the classifier (8 points)


1. Use the trained classifier to predict the classes of the *Validation_set* .

2. Print out the Accuracy, Precision, Recall, F1 score, AUROC, confusion matrix. For the matrix (except confusion matrix),keep 4 digits (such as 0.9856).

3. Create a list, named *model_performance_knn* with 5 elements, which are Accuracy, Precision, Recall, F1 score, AUROC, respectively. Create a numpy array named *cf_matrix* to store the confusion matrix.



```python
# Use the trained classifier to predict the classes of the Validation set
y_val_pred = knn_classifier.predict(X_val_normalized)

# Get unique classes to prepare for AUROC calculation
unique_classes = np.unique(np.concatenate((y_train, y_val)))
n_classes = len(unique_classes)

# For AUROC calculation, we need the probability estimates
y_val_proba = knn_classifier.predict_proba(X_val_normalized)

# Convert y_val to one-hot encoding for AUROC calculation
y_val_binarized = label_binarize(y_val, classes=unique_classes)

# Calculate performance metrics
accuracy = accuracy_score(y_val, y_val_pred)
precision = precision_score(y_val, y_val_pred, average='weighted')
recall = recall_score(y_val, y_val_pred, average='weighted')
f1 = f1_score(y_val, y_val_pred, average='weighted')

# Calculate AUROC - handling multi-class with one-vs-rest approach
if n_classes > 2:
    # One-vs-Rest approach for multi-class
    auroc = roc_auc_score(y_val_binarized, y_val_proba, multi_class='ovr', average='weighted')
else:
    # Binary classification
    auroc = roc_auc_score(y_val, y_val_proba[:, 1])

# Calculate confusion matrix
cf_matrix = confusion_matrix(y_val, y_val_pred)

# Create the model_performance_knn list with the required metrics
model_performance_knn = [accuracy, precision, recall, f1, auroc]
model_performance_knn = [round(item, 4) for item in model_performance_knn]

# Print out performance metrics with 4 decimal places
print("\nKNN Classifier Validation Results:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision (weighted): {precision:.4f}")
print(f"Recall (weighted): {recall:.4f}")
print(f"F1 Score (weighted): {f1:.4f}")
print(f"AUROC (weighted): {auroc:.4f}")
```

## TODO 6 Visulize the confusion matrix (2 points)

1. Use *seaborn* library to visualize the confusion matrix using *heatmap*. The labels should contain the counts and percentage.

```python
import seaborn as sns

plt.figure(dpi=100)

group_counts = ["{0:0.0f}".format(value) for value in cf_matrix.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in cf_matrix.flatten()/np.sum(cf_matrix)]
labels = [f"{v1}\n{v2}" for v1, v2, in zip(group_counts,group_percentages)]
labels = np.asarray(labels).reshape(5,5)
sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues')

plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')

print("\nmodel_performance_knn list:")
for i, metric_name in enumerate(["Accuracy", "Precision", "Recall", "F1 Score", "AUROC"]):
    print(f"{metric_name}: {model_performance_knn[i]:.4f}")
```
